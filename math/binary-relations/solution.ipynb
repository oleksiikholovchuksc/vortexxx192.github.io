### 1 

**Доказать, что в частично упорядоченном множестве наибольший элемент может быть только один.**


Допустим, что в частично упорядоченном множестве есть два разных наибольших элемента $x$ и $y$. По определению наибольшего элемента, должно выполняться следующее: $(x > y) \space \& \space (y > x)$. Но, так как отношение частичного порядка по определению обладает свойством антисимметричности, выходит, что $x$ и $y$ — один и тот же элемент.

### 3

**Установить биекцию между множествами $2^A$ и $A \mapsto \{0, 1\}$.**

Рассмотрим некоторое подмножество $S \space \colon \space S \in 2^A$. Рассмотрим также все элементы $x \in S$ и все элементы $y \notin S$. Значит, множество $S$ отобразим в такое отображение $f \in (A \mapsto \{0,1\})$, что $\forall x : f(x) \mapsto 1$, $\forall y : f(y) \mapsto 0$.

### 4
**Пусть $2 \mathbb{N}$ множество четных чисел. Доказать, что порядки $(2 \mathbb{N}, \leq)$ и $(\mathbb{N}, \leq)$ изоморфны.**

Если я правильно понял, что *"доказать, что порядки изоморфны"* означает "придумать биекцию, сохраняющую порядок", то определим отображение для перехода из $2 \mathbb{N}$ в $\mathbb{N}$:
$$f _{2 \mathbb{N} \mapsto \mathbb{N}} \space (x) = \frac{x}{2}$$

Такая биекция сохраняет порядок — если одно число меньше либо равно другому, то это отношение сохранится между ними и после умножения на одно и то же число.

### 5.1 

**Доказать, что результатом умножения двух линейных порядков также является линейный порядок.**

Сравнивать два элемента из $A_1 \times A_2$ будем следующим образом:
$$(a_1, a_2) \leq_{12} (b_1, b_2) \Leftrightarrow (a_2 \leq_2 b_2 \space \& \space a_2 \neq b_2) \space \lor \space (a_2 = b_2 \space \& \space a_1 \leq_1 b_1) \space \space \space (1)$$

Докажем, что порядок, полученный  обладает свойствами рефлексивности, антисимметричности, транзитивности и линейности. 

* *Рефлексивность*. Из-за того, что оба порядка обладают свойствами рефлексивности, правая сторона утверждения (1) всегда будет выполняться.
* *Антисимметричность*. Докажем, что если $(a_2 = b_2 \space \& \space a_1 \leq_1 b_1) \space \land \space (b_2 = a_2 \space \& \space b_1 \leq_1 a_1) \space (*)$, то $(a_1 = b_1 \space \land \space a_2 = b_2) \space (**)$: равенство $a_2$ и $b_2$ прямым текстом написано в предпосылке $(*)$, а равенство $a_1$ и $b_1$ следует из того, что $a_1 \leq_1 b_1$ и $b_1 \leq_1 a_1$, а отношение $P_1$ обладает свойством рефлексивности.
* *Транзитивность*. Рассмотрим пары $(a_1, a_2), (b_1, b_2), (c_1, c_2)$, для которых известно, что $(a_1, a_2) \leq_{12} (b_1, b_2)$ и $(b_1, b_2) \leq_{12} (c_1, c_2)$. Исходя из того, что изначальные порядки $P_1$ и $P_2$ также обладают свойством транзитивности, следует $(a_1, a_2) \leq_{12} (c_1, c_2)$.
* *Линейность*. Так как для сортировки двух элементов из $A_1 \times A_2$ нам не нужно ничего, кроме отношений $\leq_1$ и $\leq_2$, о которых известно, что с их помощью можно сравнить любые два элемента из соответствующих множеств, значит и любые две элемента из $A_1 \times A_2$ можно отсортировать.

### 5.2

**Пусть $\omega = (\mathbb{N}, \leq), \space 2 = (\{0, 1\}, \leq)$. Доказать, что $\omega \cong 2\omega$, но при этом $\omega \ncong \omega 2$.**

1. Установим следующую биекцию между множествами $\mathbb{N} = \{1, 2, 3, 4, \dots \}$ и $A_{\omega} = \{(0, 1), (1, 1), (0, 2), (1, 2), (0, 3), (1, 3), \dots \}$: поставим в соответствие первому элементу из $\mathbb{N}$ первый элемент из $A_{\omega}$, второму - второй, третьему - третий и т.д. Очевидно, что все условия изоморфизма будут выполняться.

2. Докажем, что $\omega \ncong \omega 2$. Рассмотрим множества $N$ и $A_{\omega 2}$: 
$$\mathbb{N} = \{1, 2, 3, 4, 5, 6, 7, \dots \},$$
$$A_{\omega 2} = \{ (1, 0), (2, 0), (3, 0), (4, 0), \dots, (1, 1), (2, 1), (3, 1), (4, 1), \dots \}$$

Возьмем две пары $(a_1, a_2), (b_1, b_2) \in A_{\omega 2} \space \colon \space a_2 = 0 \space \& \space b_2 = 1$. Для того, чтоб изоморфизм существовал, $\forall (x_1, x_2) \in A_{\omega 2} \space \colon \space (a_1, a_2) \leq (x_1, x_2) \leq (b_1, b_2)$ должно выполняться следующее: $f^{-1}((a_1, a_2)) \leq f^{-1}((x_1, x_2)) \leq f^{-1}((b_1, b_2))$. Однако элементов множества $A_{\omega 2}$ в промежутке от $(a_1, a_2)$ до $(b_1, b_2)$ бесконечно много, в то время как натуральных чисел между $f^{-1}((a_1, a_2))$ и $f^{-1}((b_1, b_2))$, следовательно искомый изоморфизм невозможен.
